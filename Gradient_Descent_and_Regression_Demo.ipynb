```python
import datasets
import numpy as np
```

```python
dataset = datasets.load_dataset('mandarjoshi/trivia_qa', 'rc')
```

```python
train_data = dataset['train']
```

```python
print(train_data.column_names)
```

```python
def_question_len = np.array([len(q) for q in train_data['question']])
def_num_ans = np.array([len(a['aliases']) for a in train_data['answer']])
def_search_res_len = np.array([len(r) for r in train_data['search_results']])
```

```python
indice = [i for i, x in enumerate(def_question_len) if x > 0 and def_num_ans[i] > 0 and def_search_res_len[i] > 0]
question_len = np.array([def_question_len[i] for i in indice])
num_ans = np.array([def_num_ans[i] for i in indice])
search_res_len = np.array([def_search_res_len[i] for i in indice])
```

```python
print(question_len)
print(num_ans)
print(search_res_len)
```

Gradied

```python
import matplotlib.pyplot as plt
```

```python
X = (question_len - question_len.mean()) / question_len.std()
X = X.reshape(-1, 1)
X = np.hstack([np.ones((X.shape[0], 1)), X])
y = num_ans
```

```python
theta = np.zeros(X.shape[1])
```

```python
alpha = 0.01
num_iter = 1000
cost = []
```

```python
for i in range(num_iter):
    gradient = X.T @ (X @ theta - y) / y.size
    theta -= alpha * gradient
    costs = (1 / (2 * y.size)) * np.sum((X @ theta - y) ** 2)
    cost.append(costs)
```

```python
print("optimal weight: ", theta)
```

```python
plt.plot(cost)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Gradient Descent')
plt.show()
```

Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
X = question_len.reshape(-1, 1)
y = num_ans
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print("intercept:", model.intercept_)
print("coefficient:", model.coef_)
```

```python
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Linear fit')
plt.xlabel('Question len')
plt.ylabel('Ans len')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

Multi

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.vstack((question_len, search_res_len)).T
y = num_ans
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(question_len, search_res_len, y, color='blue', label='Data points')
```

```python
x_surf, y_surf = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max(), 100), np.linspace(X[:,1].min(), X[:,1].max(), 100))
z_surf = model.predict(np.array([x_surf.ravel(), y_surf.ravel()]).T).reshape(x_surf.shape)
```

```python
ax.plot_surface(x_surf, y_surf, z_surf, color='red', alpha=0.5)
ax.set_xlabel('Question len')
ax.set_ylabel('Search result len')
ax.set_zlabel('Ans len')
plt.title('Multiple Linear Regression')
ax.legend()
plt.show()
```
